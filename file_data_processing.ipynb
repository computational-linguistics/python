### Read the file content using for loop and file object

```python
# uer input file is stored in a subdirectory called "data"
# the relative path is ./data/memo.txt (or data/memo.txt)
filename = input('Enter file: ')
infile = open(filename)
word_list = []
for line in infile:
    words = line.split()
    for word in words :
        if word in word_list : continue
        word_list.append(word)
        
infile.close()
print(word_list)
```

```python
# uer input file is stored in a subdirectory called "data"
# the relative path is ./data/memo.txt (or data/memo.txt)
fname = input("Enter file name: ")
infile = open(fname)
wordlist = list()
for line in infile:
    line = line.rstrip()
    words = line.split()
    for i in range(len(words)) :
        if words[i] in wordlist : continue
        wordlist.append(words[i])
        wordlist.sort()
        
infile.close()
print(wordlist)
```

### Writing to a text file

```python

filename = input('Enter filename for writing:')
outfile = open(filename, 'w')

outfile.write('This is the first line.')
outfile.write(' Still the first line... \n')
outfile.write('Now we are in the second line.\n')
outfile.write('Non string value like ' + str(5) + ' must be converted first.\n')
outfile.write('Another way: non string value like {} using format() function.\n'.format(5))

outfile.close()

print('File writing is done! Check the current directory')
```

### csv file read/write

csv files: data seperated by a comma

```python
# uer input file is stored in a subdirectory called "data"
# the relative path is ./data/beam.csv (or data/beam.csv)
import csv

filename = input('Enter filename:')
csvfile = open(filename)
with csvfile:
    reader = csv.reader(csvfile)
    print(type(reader))
    for row in reader:
        print(row)
        
csvfile.close()
```

```python
import csv

csvfile = open('names.csv', 'w')
with csvfile:
    fieldnames = ['first_name', 'last_name']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    
    print('File writing started . . . ')
    writer.writerow({'first_name': 'Baked', 'last_name': 'Beans'})
    writer.writerow({'first_name': 'Lovely', 'last_name': 'Spam'})
    writer.writerow({'first_name': 'Wonderful', 'last_name': 'Spam'})

csvfile.close()
print('File writing is done! Check the current directory')
```

### Text Processing with Unicode

Learn how to use Unicode for processing texts that use non-ASCII character sets.

Unicode supports over a million characters. Each character is assigned a number, called a code point. In Python, code points are written in the form \uXXXX, where XXXX is the number in 4-digit hexadecimal form.

```python
# NLTK's corpus files
# A snippet of Polish text (from the Polish Wikipedia; 
# see http://pl.wikipedia.org/wiki/Biblioteka_Pruska
# It is encoded as Latin-2, also known as ISO-8859-2.
import nltk

# The Python open() function can read encoded data 
# into Unicode strings, and write out Unicode strings in encoded form.
path = nltk.data.find('corpora/unicode_samples/polish-lat2.txt')
infile = open(path, encoding='latin2')

for line in infile:
    line = line.strip()
    print (line)
    
infile.close()
```
